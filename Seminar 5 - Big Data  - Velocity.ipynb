**How is data processed?**

IoT data processing involves both edge and cloud computing. At the edge, devices often have embedded processing capabilities to perform initial data analysis and filtering. Edge gateways aggregate and preprocess data before sending relevant information to the cloud for further processing. Cloud-based platforms leverage technologies like Apache Kafka, Apache Flink, and other stream processing frameworks to manage the real-time nature of IoT data.

**How data is analyzed and what was learned from that analysis?**

*The topic done in the Jupyter NoteBook(Seminar VI) serves as an example of how data velocity is managed in today's world. This example was inspired by the sentiment analysis of Facebook posts conducted by Kharwal, A. (2021)*

- Real-time processing: The surging volume and velocity of data underscore the necessity for real-time processing. Effectively overseeing substantial data volumes in near real-time poses a challenge for conventional data processing systems.
- Data integration: The diverse origins and formats of big data present integration challenges, demanding standardization, and mapping efforts to unify data into a generic format.
- Data quality: The heightened velocity of data introduces a heightened risk of errors and inconsistencies, potentially impacting the accuracy and reliability of insights derived from the data.
- Data storage: Coping with swift data velocity necessitates high-speed storage systems capable of managing large data volumes, a demand that traditional storage systems may struggle to meet.
- Security: The escalating velocity of data accentuates the need for robust security mechanisms, encompassing secure data transfer, access control, and encryption to uphold data privacy and security.
